```python
from pyspark.sql.functions import col, date_format, sum, avg, count

# 1. Read from the clean Silver table
silver_df = spark.read.table("workspace.default.silver_transactions")

# 2. Aggregate transactions to monthly customer profiles
gold_monthly_df = (silver_df
    .groupBy(
        col("customer_id"), 
        date_format(col("transaction_date"), "yyyy-MM").alias("reporting_month")
    )
    .agg(
        sum("amount").alias("total_spent"),
        avg("amount").alias("average_transaction_amount"),
        count("transaction_id").alias("transaction_count")
    )
)

# 3. Write to a Gold Delta table (overwrite mode for reporting updates)
gold_monthly_df.write.mode("overwrite").saveAsTable("workspace.default.gold_customer_monthly_summary")

print("Gold table successfully updated!")
```
